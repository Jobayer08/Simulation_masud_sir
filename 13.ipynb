```python
import random
import math
import numpy as np
import matplotlib.pyplot as plt


n = 5000
lam = 0.5


samples = []

for i in range(n):

    u = random.random()

    x = -(1 / lam) * math.log(1 - u)

    samples.append(x)



plt.figure(figsize=(10, 6))

plt.hist(
    samples,
    bins=50,
    density=True,
    alpha=0.6,
    label="Generated Samples"
)



x = np.linspace(0, max(samples), 1000)

pdf = lam * np.exp(-lam * x)

plt.plot(
    x,
    pdf,
    linewidth=2,
    label="Theoretical Exponential PDF"
)

plt.xlabel("Random Variable")
plt.ylabel("Density")
plt.title("Exponential Distribution using Inverse Transform Method")
plt.legend()
plt.grid(True)

plt.show()



print("Number of Samples =", n)
print("Lambda =", lam)
print("Sample Mean =", round(np.mean(samples), 4))
print("Theoretical Mean =", round(1 / lam, 4))
```
