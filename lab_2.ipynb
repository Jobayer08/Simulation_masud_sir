```python
import random
import matplotlib.pyplot as plt


SIM_DAYS = 100
INITIAL_INVENTORY = 50
REORDER_POINT = 20
REORDER_QUANTITY = 40

HOLDING_COST_PER_UNIT = 1
SHORTAGE_COST_PER_UNIT = 5


inventory = INITIAL_INVENTORY
holding_cost = 0
shortage_cost = 0

days = []
inventory_levels = []

for day in range(1, SIM_DAYS + 1):

    
    demand = random.randint(1, 5)

    inventory -= demand

    
    if inventory >= 0:
        holding_cost += inventory * HOLDING_COST_PER_UNIT
    else:
        shortage_cost += abs(inventory) * SHORTAGE_COST_PER_UNIT

    
    if inventory <= REORDER_POINT:
        inventory += REORDER_QUANTITY

    
    days.append(day)
    inventory_levels.append(inventory)


print("Inventory Management Simulation")
print("Initial Inventory =", INITIAL_INVENTORY)
print("Reorder Point =", REORDER_POINT)
print("Reorder Quantity =", REORDER_QUANTITY)
print("Holding Cost =", holding_cost)
print("Shortage Cost =", shortage_cost)
print("Total Cost =", holding_cost + shortage_cost)

plt.figure(figsize=(10,5))
plt.plot(days, inventory_levels, marker='o')
plt.xlabel("Day")
plt.ylabel("Inventory Level")
plt.title("Inventory Level Over Time")
plt.grid(True)
plt.show()
```
